```python
import numpy as np
import matplotlib.pyplot as plt

# Definir los parmetros
f1 = 1  # Hz
f2 = 4 * f1  # Hz

A1 = 1
A2 = 1
A3 = 4

omega1 = 2 * np.pi * f1
omega2 = 2 * np.pi * f2

phi1 = 0
phi2 = 0
phi3 = np.radians(45)  # Convertir a radianes

# Definir el tiempo
t = np.linspace(0, 2, 1000)  # 2 segundos con 1000 puntos

# Definir las funciones
S1 = A1 * np.sin(omega1 * t + phi1)
S2 = A2 * np.sin(omega2 * t + phi2)
S3 = A3 * np.sin(omega2 * t + phi3)

# Crear la grfica
plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.plot(t, S1, label="A=1, =2f1, =0")
plt.title("S(t) = A sin(t + )")
plt.ylabel("S1(t)")
plt.grid(True)
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(t, S2, label="A=1, =2f2, =0")
plt.ylabel("S2(t)")
plt.grid(True)
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(t, S3, label="A=4, =2f2, =45")
plt.ylabel("S3(t)")
plt.xlabel("Tiempo (s)")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
